```python
import pandas as pd
dataset = pd.read_csv("insurance_pre.csv")
dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
#split train and test data
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
#using GridSearchCrossValidation, no need to split/pre process the input and output.
#It will directly give best model with hyper tuning parameter.
#Model creation 
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid = {'criterion' : ('squared_error','absolute_error','friedman_mse','poisson'),
             'splitter' : ('best','random')}
regressor = GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
regressor.fit(independent,dependent)
```

```python
bestParamResult = regressor.cv_results_
print("The R_score value for best parameter {}:".format(regressor.best_params_))
```

```python
table = pd.DataFrame.from_dict(bestParamResult)
table
```

```python
finalResult = regressor.predict([[5425,34,4,0,1]])
finalResult
```
